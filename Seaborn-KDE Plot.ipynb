```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts

pd.set_option('display.max_rows',None)
```

```python
mart = pd.read_excel("C:/Users/DEV/Downloads/supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

# create a basic KDE Plot for total column

```python
 sns.kdeplot(data=mart,x = "Total")
```

# Create  KDE for all the numeric variable in dataframe

```python
sns.kdeplot(data=mart)
```

# Adjust the smooting using bw_adjust

```python
sns.kdeplot(data=mart,x='Total',bw_adjust=0.5)
```

# Group the KDE on a category variable

```python
sns.kdeplot(data=mart,x='Total',hue='Payment')
```

#    Stack KDE on a category using MULTIPLE arguement

```python
sns.kdeplot(data= mart,x='Total',hue='Payment',multiple='stack')
```

#    Use log scaling to map the variable in KDE

```python
sns.kdeplot(data=mart,x='Total',log_scale=True)
```

# change styling of hued KDE using linewidth ,palette,alpha, etc

```python
sns.kdeplot(data= mart,x='Total',hue='Payment',multiple='stack',linewidth=5,palette='Dark2',alpha=0.5)
```

# Create a bivariate KDE

```python
sns.kdeplot(data= mart,x='Unit price',y='gross income')
```

# Groups the bivariate KDE on a categorical variable and show the contours

```python
sns.kdeplot(data= mart,x='Unit price',y='gross income',hue='Gender')
```

```python
sns.kdeplot(data= mart,x='Unit price',y='gross income',hue='Gender',fill=True,levels=5,thresh=0.3)
```
