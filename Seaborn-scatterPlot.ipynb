```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
mart = pd.read_csv(r"C:\Users\DEV\Downloads\mart_train_sample.csv").sample(50)
# print(mart['Outlet_Location_Type'].unique())
# Convert the 'Outlet_Location_Type' column to a string type before splitting
mart['Outlet_Location_Type'] = mart['Outlet_Location_Type'].astype(str)
# Split the column and assign the second part (index 1) to the 'Tier' column
mart['Tier'] = mart['Outlet_Location_Type'].str.split(expand=True)[1]
# View the result
mart.head()
```

# Create a basic Scatter PLot

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales')
```

# Grouping basis on a categorical variable using HUE 

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales',hue='Outlet_Size')
```

# Grouping basis on a categorical variable using STYLE and styling Markers as well

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales',style='Outlet_Size',
               markers={"High":"^","Small":"v","Medium":"o"})
```

# Grouping basis on a categorical variable using STYLE & HUE both togather

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales',style='Outlet_Size',
                hue='Outlet_Size',
               markers={"High":"^","Small":"v","Medium":"o"})
```

# grouping basis on numeric variable  using HUE and use Palette

# 

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales',
#                 style='Outlet_Size',
                hue='Outlet_Size',
                palette='crest_r',
               markers={"High":"^","Small":"v","Medium":"o"})
```

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales',
#                 style='Outlet_Size',
                size='Tier',
                sizes=(20,200),
                hue='Outlet_Size',
                legend='full',
                palette='crest_r',
               markers={"High":"^","Small":"v","Medium":"o"})
```

# Change the Marker size with S arguement 

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales',s=500,color='red',edgecolor='black')
```
