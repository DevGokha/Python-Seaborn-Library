```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts

pd.set_option('display.max_rows',None)
```

```python
    mart = pd.read_excel("C:/Users/DEV/Downloads/supermarket_sales.xlsx")
    mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
    mart.head(10)
```

```python
sns.rugplot(data=mart,x='Unit price')
```

# Create a Rug plot for one variables 

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x='Unit price')
```

# Create a Rug plot for two variables

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x='Unit price',y='gross income')
```

# Group it by a categorical variable variable using hue 

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x='Unit price',y='gross income',hue='Gender')
```

# change in height of rug in the rug plot

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x='Unit price',y='gross income',hue='Gender',height=0.08)
```

# combine this with a KDE plot

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart,x='Unit price',y = 'gross income',hue='Gender',fill=True)
sns.rugplot(data=mart,x='Unit price',y = 'gross income',hue='Gender')
```

# combine this with a scatter plot

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=mart,x='Unit price',y = 'gross income',hue='Gender')
sns.rugplot(data=mart,x='Unit price',y = 'gross income',hue='Gender')
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart,x='Unit price',y = 'gross income',hue='Gender',fill=True)
sns.rugplot(data=mart,x='Unit price',y = 'gross income',hue='Gender',height= -0.04,clip_on=False)
```
