```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel("C:/Users/DEV/Downloads/supermarket_sales.xlsx")
mart.head()
```

```python
sns.boxplot(y='Total',data=mart,width=0.2)
```

```python
sns.boxplot(x='Payment',hue='Gender',y='Total',data=mart)
```

```python
sns.boxplot(x='Payment',hue='Gender',y='Total',data=mart,showmeans=True)
```

```python
sns.boxplot(x='Payment',hue='Gender',y='Total',data=mart,showmeans=True,meanprops={"marker":"o"
                                                                                  ,"markerfacecolor":"white"
                                                                                  ,"markersize":"15"
                                                                                  ,"markeredgecolor":"black"})
```

```python
sns.boxplot(y='Payment',x='Total',hue='Gender',data=mart,showmeans=True,meanprops={"marker":"o"
                                                                                  ,"markerfacecolor":"white"
                                                                                  ,"markersize":"15"
                                                                                  ,"markeredgecolor":"black"})
```

```python
sns.boxplot(y='Payment',x='Total',hue='Gender',data=mart,showmeans=True,meanprops={"marker":"o"
                                                                                  ,"markerfacecolor":"white"
                                                                                  ,"markersize":"15"
                                                                                  ,"markeredgecolor":"black"},palette='CMRmap',linewidth=2)
```
