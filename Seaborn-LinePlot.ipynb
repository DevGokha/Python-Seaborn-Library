```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
mart = pd.read_csv(r"C:\Users\DEV\Downloads\mart_train_sample.csv").sample(50)
# print(mart['Outlet_Location_Type'].unique())
# Convert the 'Outlet_Location_Type' column to a string type before splitting
mart['Outlet_Location_Type'] = mart['Outlet_Location_Type'].astype(str)
# Split the column and assign the second part (index 1) to the 'Tier' column
mart['Tier'] = mart['Outlet_Location_Type'].str.split(expand=True)[1]
# View the result
mart.head()
```

# create a basic line plot and try differnent estimators

```python
sns.lineplot(data=mart,x='Outlet_Year',y="Sales",ci=None,estimator=None)
```

```python
#Estimator=sum/None
height = [5,6,7,5,8]
age = [22,19,23,26,33]
sns.lineplot(x=age,y=height,estimator=sum)
```

# Testing with dfferent confidence Intervals and with different number of bootstraping

```python
sns.lineplot(data=mart,x='Outlet_Year',y="Sales",ci='sd')
```

# Grouping using HUE and use different palettes

```python
sns.lineplot(data=mart,x='Outlet_Year',y="Sales",ci=None,hue='Outlet_Size'
            ,palette=['red','green','blue'])
```

# grouping using style,use different Marker and Dashes

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",ci=None,style='Outlet_Size'
            ,markers=True,
            dashes=False
            )
```

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",ci=None
            ,size='Outlet_Size' )
```

# Use Different semantic styling parameters on same or different variables

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",ci=None,
             hue="Outlet_Size"
            ,style='Tier',
            markers=True)
```
