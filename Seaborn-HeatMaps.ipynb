```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel("C:/Users/DEV/Downloads/mart_linePlot (1).xlsx")
mart.columns = mart.columns.str.lower()
mart.head(10)
```

```python
mart.shape
```

```python
martPiv = mart.pivot_table(index='outlet_year',columns='outlet_size',values='sales')
martPiv
```

```python
sns.set_style()
plt.figure(figsize=(5,5))
sns.heatmap(martPiv,
#             square=True
           annot=True,
            fmt='.0f',
            annot_kws = dict(size=15,weight='bold'),
            linewidths=0.5,
            linecolor='black',
            cmap='OrRd_r')
```

```python
numeric_mart = mart.select_dtypes(include=[np.number])
sns.heatmap(numeric_mart.corr(),
            vmin=-1, vmax=1, center=0, cmap='RdYlBu_r', annot=True, fmt='.1f', annot_kws={'size':15, 'weight':'bold'}, linecolor='black')
```
