```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts

pd.set_option('display.max_rows',None)
```

```python
mart = pd.read_excel("C:/Users/DEV/Downloads/supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

# basic histogram


```python
sns.histplot(data=mart,x='Total')
```

# change the bins width using binwidth arguement

```python
sns.histplot(data=mart,x='Total',binwidth=130)
```

# change number of Bins and interval

```python
plt.figure(figsize=(15,5))
sns.histplot(data=mart,x='Total',bins=np.arange(0,1100,50))
plt.xticks(np.arange(0,1100,50))
```

# combine with KDE

```python

sns.histplot(data=mart,x='Total',kde=True)
```

# use a catigorical variable in HUE and STACK it using MULTIPLE argrument

```python
sns.histplot(data=mart,x='Total',hue='Payment',multiple='stack')
```

# make it a step/poly plot using ELEMENT arguement and change the FILL

```python
sns.histplot(data=mart,x='Total',element='step',fill=False)
```

# use categorical variable an shrink it 

```python
# sns.histplot(data=mart,x='Payment',shrink=0.7)
sns.histplot(data=mart,x='Total',stat='probability')
```

# create a Bivariate Histogram

```python
sns.histplot(data=mart,x='Total',y ='gross income')
```
