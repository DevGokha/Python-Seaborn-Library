```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#drawing a first basic joint in plot
sns.jointplot(data=iris,x='sepal_length',y='petal_length')
```

# change its kind to "scatter"|"kde"|"hist"|"hex"|"reg"|"resid"

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",kind="reg")
```

# grouping based on a categorical variable

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",hue='species')
```

# Formatting all and individual Plots

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",
#               color='red',
#              hue='species',
 #             palette="BuPu",
             joint_kws=dict(marker='+',color='red'),
             marginal_kws=dict(color='green',kde=True))
```

# Adjust height , ratio,space,and show/hide the marginal_ticks

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length"
              ,height=4
             ,ratio=2
#              ,space=0.1
             ,marginal_ticks=True)
```

```python
l=sns.jointplot(data=iris,x="sepal_length",y="petal_length",joint_kws=dict(color='red'))
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot,height=0.2)
```
